```python
# Import requests and BeautifulSoup (if not already imported)
import requests
from bs4 import BeautifulSoup
import pandas as pd

# GameStop Revenue URL
url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(html_gme, "html.parser")

# Create dataframe
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Extract revenue table
for row in soup_gme.find_all("tr"):
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text
        revenue = cols[1].text.replace("$","").replace(",","")
        if revenue != "":
            gme_revenue = gme_revenue.append({"Date": date, "Revenue": revenue}, ignore_index=True)

# Remove missing data
gme_revenue.dropna(inplace=True)

# Display last 5 rows
gme_revenue.tail()
```
